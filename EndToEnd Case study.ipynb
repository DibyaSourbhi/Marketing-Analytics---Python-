## 2. Performing Customers/sellers Segmentation

```python
# Use merged datasets of item_order and prod
item_prod.columns
```

```python
item_prod.Total_price
```

```python
# Find out minimum and maximum value for creating bin
print(item_prod.Total_price.max())
print(item_prod.Total_price.min())
```

```python
revenue_customer=item_prod.groupby('customer_id')['Total_price'].sum().reset_index()
revenue_customer
```

```python
# Make bins for grouping
revenue_customer['revenue_group'] = pd.cut(revenue_customer.Total_price, bins= np.arange(0 , 7000, 500), labels=['0-500','500-1000','1000-1500','1500-2000', '2000-2500','2500-3000','3000-3500','3500-4000','4000-4500','4500-5000','5000-5500','5500-6000','6000-6500'])
```

```python
revenue_customer
```

```python
count_of_customer=revenue_customer.value_counts('revenue_group')
count_of_customer
```

```python
count_of_customer.plot(kind='line', figsize=(10,8), color='g')
```
